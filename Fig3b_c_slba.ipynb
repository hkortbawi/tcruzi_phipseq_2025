# Fig 3b and c - SLBA results

```python
ts2_23 = pd.read_csv(dir+"fig3b_slba.csv",index_col=False)
ala_scan = pd.read_csv(dir+"fig3c_slba.csv", index_col=False)
ts2_23.head()
```

```python
# plot slba data
sns.set_style('ticks')

order = ['seroneg', 'seropos']

ax = sns.stripplot(data=ts2_23, x='sero', y='ts_2.23_norm_ab_index', size = 6,
                   order=order,
                   linewidth=1
                  )

ax.set_title("")

plt.rcParams["figure.figsize"] = (1, 2)
plt.rcParams['axes.linewidth'] = .75

ax.set_xlabel("")
ax.set_ylabel("Antibody Index")

plt.savefig(fig_dir+'fig_3a.pdf', dpi=300 , bbox_inches="tight")
plt.show()
```
